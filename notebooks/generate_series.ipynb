```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
observations = 120
```

```python
l = np.linspace(0, 3, observations)
i = np.random.rand(observations)
i = i + l
j = np.array(([1, 5, 5, 1, 0, 0, 1, -3, -3, 1, 0, 8] * int(observations/12)))
j = j / 2.5
j = j * np.random.rand(observations)
```

```python
i_S = pd.Series(i)
i_S.plot(kind='line')
```

```python
j_S = pd.Series(j)
j_S.plot(kind='line')
```

```python
k = i + j
```

```python
k_S = pd.Series(k)
```

```python
k_S.plot(kind='line')
```
